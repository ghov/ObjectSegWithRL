```python
#from os import listdir
from torchvision.transforms import Resize
from PIL import Image
import skimage.io as io
%matplotlib inline
from pycocotools.coco import COCO
from os import listdir
import matplotlib.pyplot as plt
import numpy as np
```

```python
read_directory = '/media/greghovhannisyan/BackupData1/mscoco/images/train2017/'
#write_directory = '/media/greghovhannisyan/BackupData1/mscoco/images/resized_images/train_2017_224_224/'
annFile = '/media/greghovhannisyan/BackupData1/mscoco/annotations/instances/instances_train2017.json'
bbox_crop_write_dir = '/media/greghovhannisyan/BackupData1/mscoco/images/train2017_crop_bbox/'
```

```python
# initialize COCO api for instance annotations
coco=COCO(annFile)
```

```python
# Put the data into a dictionary, with each ['id'] being key
# The values are the lenth of the list and the length of the polygons in the list
id_dict = dict()


for filename in listdir(read_directory):
    #print(filename.split('.')[0])
    annIds = coco.getAnnIds(imgIds=int(filename.split('.')[0]), iscrowd=None)
    #print(annIds)
    anns = coco.loadAnns(annIds)
    #print(anns)
    for ann in anns:
        if(ann['id'] in id_dict):
            print('error: the same id showed up twice.')
        else:
            id_dict[ann['id']] = dict()
            id_dict[ann['id']]['polygons'] = len(ann['segmentation'])
            if(len(ann['segmentation']) == 0):
                id_dict[ann['id']]['vertices'] = 0
                print('error: no polygon.')
            elif(len(ann['segmentation']) == 1):
                id_dict[ann['id']]['vertices'] = len(ann['segmentation'][0])
            elif(len(ann['segmentation']) > 1):
                id_dict[ann['id']]['vertices'] = list()
                for poly in ann['segmentation']:
                    id_dict[ann['id']]['vertices'].append(len(poly))
        
        #print(type(val['segmentation']))
        #print(len(val['segmentation']))
        #print(val['segmentation'])
```

```python
# Count how many have only one polygon
count = 0
poly_size_list = list()
for key, val in id_dict.items():
    if(id_dict[key]['polygons'] == 1):
        count += 1
        poly_size_list.append(id_dict[key]['vertices'])
print(count)
```

```python
size_count_dict = dict()
poly_size_set = set(poly_size_list)
for val in poly_size_set:
    size_count_dict[val] = poly_size_list.count(val)
```

```python
# make a dictinary, where the key is the size of the polygon and the value is a list of segmentation ids
# The items in this dict will only have one polygon

vertex_count_dict = dict()
for key, val in id_dict.items():
    if(id_dict[key]['polygons'] == 1):-
        if(id_dict[key]['vertices'] in vertex_count_dict):
            # Add to the existing list
            vertex_count_dict[id_dict[key]['vertices']].append(key)
        else:
            # make a new list
            vertex_count_dict[id_dict[key]['vertices']] = list()
            vertex_count_dict[id_dict[key]['vertices']].append(key)
```

```python
# Get a list of all of the image ids that have only one polygon for each annotation.
one_polygon_img_ids = set()

for filename in listdir(read_directory):
    #print(filename.split('.')[0])
    annIds = coco.getAnnIds(imgIds=int(filename.split('.')[0]), iscrowd=None)
    #print(annIds)
    anns = coco.loadAnns(annIds)
    #print(anns)
    for ann in anns:
        if(len(ann['segmentation']) == 1):
            one_polygon_img_ids.add(filename.split('.')[0])
            
        #print(type(val['segmentation']))
        #print(len(val['segmentation']))
        #print(val['segmentation'])
```

```python
len(one_polygon_img_ids)
```

```python
# for each id in one_polygon_img_ids, open the image.
# Then get the segmentation for said image.
# Then use the bounding boxes to crop out the boxes and save them as individual jpg file.
# The new file names will be the segmentation ids
#bbox_crop_write_dir = '/media/greghovhannisyan/BackupData1/mscoco/images/train2017_crop_bbox/'
failed_list = list()

for val in one_polygon_img_ids:
    # open image
    I = io.imread(read_directory + val + '.jpg')
    #plt.imshow(I)
    # Get the annotations
    annIds = coco.getAnnIds(imgIds=int(val), iscrowd=None)
    anns = coco.loadAnns(annIds)
    for ann in anns:
        try:
            bbox = ann['bbox']
            #print(bbox)
            x1 = int(bbox[0])
            x2 = int(bbox[0] + bbox[2])
            y1 = int(bbox[1])
            y2 = int(bbox[1] + bbox[3])
            temp_crop_arr = I[y1:y2, x1:x2]
            #plt.imshow(temp_crop_arr)
            io.imsave(bbox_crop_write_dir + str(ann['id']) + '.jpg', temp_crop_arr)

            #temp_crop_img = Image.fromarray(temp_crop_arr).convert('RGB')
            #temp_crop_img.save(bbox_crop_write_dir + str(ann['id']) + '.jpg')
            #temp_crop.save(bbox_crop_write_dir + ann['id'] + '.jpg')
        except:
            failed_list.append(ann)
```

```python
sample_ann_id = 1833954
temp_list = list()
temp_list.append(sample_ann_id)
sample_ann = coco.loadAnns(temp_list)
print(sample_ann[0]['bbox'])
print(sample_ann[0]['segmentation'][0])

xmin = sample_ann[0]['bbox'][0]
ymin = sample_ann[0]['bbox'][1]
temp_seg = sample_ann[0]['segmentation'][0]

new_seg = list()
length = len(temp_seg)
print(length)

for i in range(0,length, 2):
    new_seg.append(temp_seg[i]-xmin)
    new_seg.append(temp_seg[i+1]-ymin)
    
#print(temp_list)
#print(annIds)
```

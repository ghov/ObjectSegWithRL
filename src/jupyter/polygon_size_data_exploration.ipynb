```python
#from os import listdir
from torchvision.transforms import Resize
from PIL import Image
%matplotlib inline
from pycocotools.coco import COCO
from os import listdir
```

```python
read_directory = '/media/greghovhannisyan/BackupData1/mscoco/images/train2017/'
#write_directory = '/media/greghovhannisyan/BackupData1/mscoco/images/resized_images/train_2017_224_224/'
annFile = '/media/greghovhannisyan/BackupData1/mscoco/annotations/instances_train2017.json'
```

```python
# initialize COCO api for instance annotations
coco=COCO(annFile)
```

```python
# Put the data into a dictionary, with each ['id'] being key
# The values are the lenth of the list and the length of the polygons in the list
id_dict = dict()


for filename in listdir(read_directory):
    #print(filename.split('.')[0])
    annIds = coco.getAnnIds(imgIds=int(filename.split('.')[0]), iscrowd=None)
    #print(annIds)
    anns = coco.loadAnns(annIds)
    #print(anns)
    for ann in anns:
        if(ann['id'] in id_dict):
            print('error: the same id showed up twice.')
        else:
            id_dict[ann['id']] = dict()
            id_dict[ann['id']]['polygons'] = len(ann['segmentation'])
            if(len(ann['segmentation']) == 0):
                id_dict[ann['id']]['vertices'] = 0
                print('error: no polygon.')
            elif(len(ann['segmentation']) == 1):
                id_dict[ann['id']]['vertices'] = len(ann['segmentation'][0])
            elif(len(ann['segmentation']) > 1):
                id_dict[ann['id']]['vertices'] = list()
                for poly in ann['segmentation']:
                    id_dict[ann['id']]['vertices'].append(len(poly))
        
        #print(type(val['segmentation']))
        #print(len(val['segmentation']))
        #print(val['segmentation'])
```

```python
# Count how many have only one polygon
count = 0
poly_size_list = list()
for key, val in id_dict.items():
    if(id_dict[key]['polygons'] == 1):
        count += 1
        poly_size_list.append(id_dict[key]['vertices'])
print(count)
```

```python
size_count_dict = dict()
poly_size_set = set(poly_size_list)
for val in poly_size_set:
    size_count_dict[val] = poly_size_list.count(val)
```

```python
# make a dictinary, where the key is the size of the polygon and the value is a list of segmentation ids
# The items in this dict will only have one polygon

vertex_count_dict = dict()
for key, val in id_dict.items():
    if(id_dict[key]['polygons'] == 1):
        if(id_dict[key]['vertices'] in vertex_count_dict):
            # Add to the existing list
            vertex_count_dict[id_dict[key]['vertices']].append(key)
        else:
            # make a new list
            vertex_count_dict[id_dict[key]['vertices']] = list()
            vertex_count_dict[id_dict[key]['vertices']].append(key)
```
